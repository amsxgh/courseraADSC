Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
Q1 = df['ConvertedComp'].quantile(0.25)
Q3 = df['ConvertedComp'].quantile(0.75)
IQR = Q3 - Q1
print("First Quartile: ", Q1)
print("Third Quartile: ", Q3)
print("Inter Quartile Range: ", IQR)
```

Find out the upper and lower bounds.


```python
# your code goes here
upper_bound = Q3 + (1.5*IQR)
lower_bound = Q1 - (1.5*IQR)

print("Upper Boundary: ", upper_bound)
print("Lower Boundary: ", lower_bound)
```

```python
count = df[df['ConvertedComp'] >= 209698.0].shape[0]

print(count)
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
outliers = df[(df['ConvertedComp'] <= lower_bound) | (df['ConvertedComp'] >= upper_bound)]

num_outliers = outliers.shape[0]

print("The Number of Outliers: ", num_outliers)
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
# your code goes here
df_no_outliers = df[(df['ConvertedComp'] >= lower_bound) & (df['ConvertedComp'] <= upper_bound)]
# lower = df[df['ConvertedComp'] <= lower_bound]
# upper = df[df['ConvertedComp'] >= upper_bound]

df_no_outliers
# upper['ConvertedComp']
```
